```python
import numpy as np
import os.path as op
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
DATADIR = '/Users/Light/datascience/datasets/faces/'
TRAINING_SET = op.join(DATADIR, 'UTKFace_train')
VALIDATION_SET = op.join(DATADIR, 'UTKFAce_val')

training_filenames = os.listdir(TRAINING_SET)
validation_filenames = os.listdir(VALIDATION_SET)
print('Training set :', training_filenames.__len__())
print('Validation set :', validation_filenames.__len__())
```

```python
fig, axs = plt.subplots(3, 3, figsize=(20, 20))
axs = np.ravel(axs)

for i in range(9):
    name = training_filenames[i * 15]
    fn = op.join(DATADIR, 'UTKFace_train', name)
    img = plt.imread(fn)
    imgplot = axs[i].imshow(img)
    axs[i].set_xlabel(name)
plt.show()
```
